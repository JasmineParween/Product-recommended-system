```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('../datasets/flipkart_com-ecommerce_sample.csv')
df = pd.read_csv('../datasets/cleaned_data.csv')
```

```python
dataset.head(10)
```

```python
dataset['product_category_tree'] = dataset['product_category_tree'].replace(to_replace='["\[\]]',value="",regex=True)
```

```python
dataset['product_category_tree']
```

```python
Categories = set()
```

```python
for i in dataset['product_category_tree']:
    Categories.add(i.split(" >>")[0])
```

```python
Categories
```

```python
d = {}
```

```python
for i in dataset['product_category_tree']:
    k = i.split(" >>")[0]
    if d.get(k) == None:
        d[k] = 1
    else:
        d[k]+=1
```

```python
to_be_popped = []
for k,v in d.items():
    if v <30:
        to_be_popped.append(k)
```

```python
for i in to_be_popped:
    d.pop(i)
```

```python
d
```

```python
import json
```

```python
out_file = open("categories.json", "w")
```

```python
json.dump(d,out_file,indent=4)
```
